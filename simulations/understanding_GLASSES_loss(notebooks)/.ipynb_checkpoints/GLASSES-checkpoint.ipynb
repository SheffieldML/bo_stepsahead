```python
# This is a notebook to illustrate how the GLASSES loss function looks like
```

```python
import GPy
import numpy as np
import GPyOpt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# --- Objective function
objective_true  = GPyOpt.fmodels.experiments2d.sixhumpcamel()             # true function
objective_noisy = GPyOpt.fmodels.experiments2d.sixhumpcamel(sd = 0.1)     # noisy version
bounds = objective_noisy.bounds                                           # problem constrains 
input_dim = len(bounds)

# Select an specific kernel from GPy
kernel = GPy.kern.RBF(input_dim, variance=.1, lengthscale=.1) + GPy.kern.Bias(input_dim) # we add a bias kernel


# --- Problem definition and optimization
BO_demo_2d = GPyOpt.methods.BayesianOptimization(f=objective_noisy.f,  # function to optimize       
                                        kernel = kernel,               # pre-specified model
                                        bounds=bounds,                 # box-constrains of the problem
                                        acquisition='LCB',             # Selects the Expected improvement
                                        acquisition_par = 2,           # parameter of the acquisition function
                                        numdata_inital_design = 10,     # 15 initial points   
                                        type_initial_design='latin',   # latin desing of the initial points 
                                        model_optimize_interval= 2,    # The model is updated every two points are collected
                                        normalize = True)              # normalized y                       
```

```python
objective_true.plot()
```

```python
model = BO_demo_2d.model
model.optimize_restarts(10)
model.plot()
```

```python
## We start by plotting the expected loss
expected_loss = GPyOpt.core.acquisition.AcquisitionEL1()
expected_loss.set_model(model)
```

```python
n_grid = 50
X1 = np.linspace(bounds[0][0], bounds[0][1], n_grid)
X2 = np.linspace(bounds[1][0], bounds[1][1], n_grid)
x1, x2 = np.meshgrid(X1, X2)
X = np.hstack((x1.reshape(n_grid*n_grid,1),x2.reshape(n_grid*n_grid,1)))

## 
plt.figure(figsize=(5,5))            
plt.contourf(X1, X2, expected_loss.acquisition_function(X).reshape((n_grid,n_grid)),100)
plt.plot(model.X[:,0], model.X[:,1], 'r.', markersize=10, label=u'Observations')
plt.colorbar()  
plt.xlabel('X1')
plt.ylabel('X2')            
plt.title('Expected loss')
plt.axis((bounds[0][0],bounds[0][1],bounds[1][0],bounds[1][1]))
```

```python
# Now we set up 5 steps ahead, and we predict the locations using a batch design based on Lipschitz inference.
```

```python
def predict_locations(x_star, bounds, acquisition, n_ahead, L, Min):
    '''
    Predicts future location evaluations based on a Lipschitz inference criterion
    '''
    ### --- load the recursive method to find the batches
    from GPyOpt.core.acquisition import AcquisitionMP
    from GPyOpt.core.optimization import wrapper_DIRECT

    penalized_loss = AcquisitionMP(expected_loss,transform='softplus')
    penalized_loss.set_model(model)

    ## --- initialize the batch with the putative input
    k = 1
    X_batch = x_star

    ## --- predict the remaining future locations
    while k<n_ahead:
        penalized_loss.update_batches(X_batch,L,Min)
        new_sample = wrapper_DIRECT(penalized_loss.acquisition_function,bounds)
        X_batch = np.vstack((X_batch,new_sample))
        k +=1
    return X_batch
```

```python
def loss_GLASSES(X_star, bounds, acquisition,n_ahead,L, Min, model):
    from GPyOptmsa.quadrature.emin_epmgp import emin_epmgp
    '''
    Computes the loss n steps n_ahead for the location x_star
    '''
    # number of points to evaluate
    n_points = X_star.shape[0]
    X_loss   = np.zeros((n_points,1))
    
    ## --- Current best
    Y               = model.Y
    eta             = Y.min()

    for i in range(n_points):
        # --- Compute future locations
        future_locations = predict_locations(X_star[i,:],bounds,acquisition,n_ahead, L, Min)

        # --- Evaluate GP at the sample and compute full covariance
        m, K       = model.predict(future_locations,full_cov=True)

        # --- Compute the expected loss
        X_loss[i,:]       = emin_epmgp(m,K,eta)

    return X_loss
```

```python
## set initial point x* where we will be avaluating the function
from GPyOpt.core.optimization import estimate_L, estimate_Min
x_star = np.array([[0,0]])
n_ahead = 5
L = estimate_L(model,bounds)
Min = estimate_Min(model,bounds)
predicted_locations = predict_locations(x_star,bounds,expected_loss,n_ahead,L,Min)
```

```python
## plot of the penalized loss
from GPyOpt.core.acquisition import AcquisitionMP
penalized_loss = AcquisitionMP(expected_loss,transform='softplus')
penalized_loss.set_model(model)
penalized_loss.update_batches(predicted_locations,L,Min)

plt.figure(figsize=(5,5))            
plt.contourf(X1, X2, penalized_loss.acquisition_function(X).reshape((n_grid,n_grid)),100)
plt.plot(model.X[:,0], model.X[:,1], 'r.', markersize=10, label=u'Observations')
plt.plot(predicted_locations[:,0], predicted_locations[:,1], 'w+', markersize=10, label=u'Observations')
plt.colorbar()  
plt.xlabel('X1')
plt.ylabel('X2')            
plt.title('Expected loss')
plt.axis((bounds[0][0],bounds[0][1],bounds[1][0],bounds[1][1]))
```

```python
loss_glasses = loss_GLASSES(X, bounds, expected_loss,n_ahead,L, Min, model)
```

```python
loss_glasses = loss_GLASSES(X, bounds, expected_loss,n_ahead,L, Min, model)
plt.figure(figsize=(5,5))            
plt.contourf(X1, X2, loss_glasses.reshape((n_grid,n_grid)),100)
plt.plot(model.X[:,0], model.X[:,1], 'r.', markersize=10, label=u'Observations')
plt.colorbar()  
plt.xlabel('X1')
plt.ylabel('X2')            
plt.title('Expected loss')
plt.axis((bounds[0][0],bounds[0][1],bounds[1][0],bounds[1][1]))
```
